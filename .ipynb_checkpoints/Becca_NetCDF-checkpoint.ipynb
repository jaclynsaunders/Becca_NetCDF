```python
### Convert coordinates to decimal degrees
### Assumes Lat in deg N and Lon in deg W

def Cart_Coord(deg, min):
    i = 0
    newCart = []
    while i < len(deg):
        DD = deg[i] + (min[i]/60)
        i+=1
        newCart.append(DD)
        
    return newCart
```
